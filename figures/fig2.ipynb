```python
import wandb
import numpy as np
import matplotlib.pyplot as plt
import jax.numpy as jnp
from jax import grad, jit, vmap
from jax import random
import seaborn as sns

api = wandb.Api()

custom_CMAP = sns.cubehelix_palette(8, start=0.5, rot=-0.75, as_cmap=True, reverse=True)
colors = custom_CMAP(jnp.linspace(0, 1, 7))
fontsize = 25
save_every = 1
```

```python
run = api.run("/varadi-epfl/ICL/runs/bvqib1kg") #fig2
# run = api.run("/varadi-epfl/ICL/runs/szvmyce2") #transformer-onehot
# run = api.run("/varadi-epfl/ICL/runs/rjwykcrc") #transformer-learned

h = run.scan_history()

losses = [row["loss"] for row in h]
```

```python
fig = plt.figure(figsize=(8, 12))
plt.plot(jnp.arange(len(losses)) * save_every, jnp.array(losses), c=colors[0], zorder=1)
plt.scatter(115, 1.668,c='r', s=20)
plt.text(205,1.669,"$t = 115$",rotation=0, fontsize=fontsize * 0.9)
plt.scatter(5000, 1.6355,c='r', s=50 )
plt.text(5095,1.638,'$t = 5000$',rotation=0, fontsize=fontsize * 0.9)
plt.scatter(16000, 1.6068,c='r' , s=50, zorder=5)
plt.text(13000,1.61,'$t = 16000$',rotation=0, fontsize=fontsize * 0.9)
plt.legend(fontsize=fontsize * 0.8)
sns.despine()
plt.xlabel("Step $(t)$", fontsize=fontsize)
plt.ylabel("Loss", fontsize=fontsize)
plt.yticks(fontsize=fontsize * 0.6)
plt.xticks(fontsize=fontsize * 0.6)

plt.show()
```
